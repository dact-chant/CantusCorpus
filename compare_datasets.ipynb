### Code for basic comparing different versions of CantusCorpus dataset

```python
import pandas as pd
```

```python
CC_1_CHANTS = 'cantuscorpus_1.0/cantuscorpus_1.0/chants.csv'
CC_2_CHANTS = 'cantuscospus_2.0/cantuscorpus_2.0/chants.csv'
CC_1_SOURCES = 'cantuscorpus_1.0/cantuscorpus_1.0/sources.csv'
CC_2_SOURCES_SCRAPED = 'cantuscospus_2.0/cantuscospus_2.0/scraped_sources.csv'
CC_2_SOURCES_MISSING = 'cantuscospus_2.0/cantuscospus_2.0/missing_sources.csv'
```

```python
v1_chants = pd.read_csv(CC_1_CHANTS)
v2_chants = pd.read_csv(CC_2_CHANTS)

v1_sources = pd.read_csv(CC_1_SOURCES)
v2_sources_scraped = pd.read_csv(CC_2_SOURCES_SCRAPED)
v2_sources_missing = pd.read_csv(CC_2_SOURCES_MISSING)
```

```python
print(f'Number of chants in CC 1.0: {len(v1_chants)}')
print(f'Number of chants in CC 2.0: {len(v2_chants)}')
print(f'Difference in number of chants: {len(v2_chants) - len(v1_chants)}')
print('---------------------------------')
print(f'Number of sources in CC 1.0: {len(v1_sources)}')
print(f'Number of sources in CC 2.0 (scraped): {len(v2_sources_scraped)}')
print(f'Difference in number of sources: {len(v2_sources_scraped) - len(v1_sources)}')
```

```python
# Inspect difference based on databases
databases = v1_chants['db'].unique()
for db in databases:
    v1_db_chants = v1_chants[v1_chants['db'] == db]
    v2_db_chants = v2_chants[v2_chants['db'] == db]
    print(f'Database: {db}')
    print(f'  Number of chants in CC 1.0: {len(v1_db_chants)}')
    print(f'  Number of chants in CC 2.0: {len(v2_db_chants)}')
    print(f'  Difference in number of chants: {len(v2_db_chants) - len(v1_db_chants)}')
```

```python
# Inspect difference based on main genres
MAIN_GENRES = ['A', 'V', 'R', 'Tp', 'W']
for genre in MAIN_GENRES:
    v1_genre_chants = v1_chants[v1_chants['genre'] == genre]
    v2_genre_chants = v2_chants[v2_chants['genre'] == genre]
    print(f'Genre: {genre}')
    print(f'  Number of chants in CC 1.0: {len(v1_genre_chants)}')
    print(f'  Number of chants in CC 2.0: {len(v2_genre_chants)}')
    print(f'  Difference in number of chants: {len(v2_genre_chants) - len(v1_genre_chants)}')
```

```python
# Inspect abnormal or very high expansions in genre
genres = v1_chants['genre'].unique()
for genre in genres:
    v1_genre_chants = v1_chants[v1_chants['genre'] == genre]
    v2_genre_chants = v2_chants[v2_chants['genre'] == genre]
    if (len(v2_genre_chants) > len(v1_genre_chants) * 1.5) or (len(v2_genre_chants) - len(v1_genre_chants) > 2000):  # Arbitrary threshold for "abnormal" expansion
        print(f'Genre: {genre}')
        print(f'  Number of chants in CC 1.0: {len(v1_genre_chants)}')
        print(f'  Number of chants in CC 2.0: {len(v2_genre_chants)}')
        print(f'  Difference in number of chants: {len(v2_genre_chants) - len(v1_genre_chants)}')
```

```python
# Inspect decrease by genre
i = 0
for genre in genres:
    v1_genre_chants = v1_chants[v1_chants['genre'] == genre]
    v2_genre_chants = v2_chants[v2_chants['genre'] == genre]
    if len(v2_genre_chants) < len(v1_genre_chants):
        print(f'Genre: {genre}')
        print(f'  CC 1.0: {len(v1_genre_chants)}, CC 2.0: {len(v2_genre_chants)}, Diff: {len(v2_genre_chants) - len(v1_genre_chants)}')
        i += 1
        
print('---------------------------------')
print('Number of genre with decrease:', i)
```

```python
# Sources in CC 1.0 not in CC 2.0
sources_1_set = set(v1_sources['srclink'])
sources_2_set = set(v2_sources_scraped['srclink'])
missing_sources = sources_1_set - sources_2_set
print(f'Number of sources in CC 1.0 not in CC 2.0 scraped: {len(missing_sources)}')
sources_2_missing = set(v2_sources_missing['srclink'])
missing_sources_final = missing_sources - sources_2_missing
print(f'Number of sources in CC 1.0 not in CC 2.0 (after accounting for known missing): {len(missing_sources_final)}')
```

```python
# Rewrite 'cantusplanus' -> 'cantus.ispan' in srclinks of missing sources and check if they are in CC 2.0
i = 0
modified_found = []
for src in missing_sources_final:
    if 'cantusplanus' in src:
        modified_src = src.replace('cantusplanus', 'cantus.ispan')
        if modified_src in sources_2_set:
            print(f'Found modified source in CC 2.0: {modified_src}')
            modified_found.append(src)
            i += 1
print(f'\nNumber of missing sources found with modified srclink: {i}')
print('\nSources not found even after modification:', set(missing_sources_final) - set(modified_found))
```
